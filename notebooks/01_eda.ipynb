```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/dataset.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
```

```python
categoricalColumns = df.select_dtypes(include="object").columns.tolist()
numericalColumns = df.select_dtypes(exclude="object").columns.tolist()

categoricalColumns, numericalColumns
```

```python
TARGET_COL = "Attrition"

DROP_COLS = [
    "EmployeeCount",
    "EmployeeNumber",
    "Over18",
    "StandardHours"
]

CATEGORICAL_FEATURES = [
    "BusinessTravel",
    "Department",
    "EducationField",
    "Gender",
    "JobRole",
    "MaritalStatus",
    "OverTime"
]

NUMERIC_FEATURES = []
for col in df.columns:
    if col not in CATEGORICAL_FEATURES and col not in DROP_COLS and col != TARGET_COL:
        NUMERIC_FEATURES.append(col)

df[NUMERIC_FEATURES]
```

```python
x = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL].map({"Yes":1, "No":0})

x = x.drop(columns=DROP_COLS)
```

```python
x.shape, y.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
# For Linear Models
numericTransformerScaling = Pipeline(steps = [
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
# For Tree Models(They Dont Require Feature Scaling)
numericTransformerNonScaling = Pipeline(steps = [
    ("imputer", SimpleImputer(strategy="median"))
])
```

```python
categoricalTransformer = Pipeline(steps = [
    ("imputer", SimpleImputer(strategy= "most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocessingLinear = ColumnTransformer(
    transformers=[
        ("num", numericTransformerScaling, NUMERIC_FEATURES),
        ("cat", categoricalTransformer, CATEGORICAL_FEATURES)
    ]
)
```

```python
preprocessingNonLinear = ColumnTransformer(
    transformers=[
        ("num", numericTransformerNonScaling, NUMERIC_FEATURES),
        ("cat", categoricalTransformer, CATEGORICAL_FEATURES)
    ]
)
```

```python
import mlflow
import mlflow.sklearn
```

```python
RANDOM_STATE = 42
EXPERIMENT_NAME = "HR_Attrition_Classification"
```

```python
mlflow.set_experiment(EXPERIMENT_NAME)
```

```python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(
    x,
    y,
    test_size=0.2,
    stratify=y,
    random_state= RANDOM_STATE
)

xTrain.shape, xTest.shape , yTrain.shape, yTest.shape
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
from sklearn.linear_model import LogisticRegression
```

```python
with mlflow.start_run(run_name="logistic_baseline"):
    model = LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        random_state= RANDOM_STATE
    )

    pipeline = Pipeline(steps = [
        ("preprocessing", preprocessingLinear),
        ("model", model)
    ])

    pipeline.fit(xTrain, yTrain)

    yPredictions = pipeline.predict(xTest)
    yProbs = pipeline.predict_proba(xTest)[:, 1]

    mlflow.log_params(model.get_params())
    
    mlflow.log_metrics({
        "accuracy": accuracy_score(yTest, yPredictions),
        "precision": precision_score(yTest, yPredictions),
        "recall": recall_score(yTest, yPredictions),
        "f1": f1_score(yTest, yPredictions),
        "roc_auc": roc_auc_score(yTest, yProbs)
    })
```

```python
with mlflow.start_run(run_name="logistic_l2_c_0.5"):
    model = LogisticRegression(
        max_iter=1000,
        C = 0.5,
        penalty= "l2",
        class_weight="balanced",
        random_state= RANDOM_STATE
    )

    pipeline = Pipeline(steps = [
        ("preprocessing", preprocessingLinear),
        ("model", model)
    ])

    pipeline.fit(xTrain, yTrain)

    yPredictions = pipeline.predict(xTest)
    yProbs = pipeline.predict_proba(xTest)[:, 1]

    mlflow.log_params(model.get_params())
    
    mlflow.log_metrics({
        "accuracy": accuracy_score(yTest, yPredictions),
        "precision": precision_score(yTest, yPredictions),
        "recall": recall_score(yTest, yPredictions),
        "f1": f1_score(yTest, yPredictions),
        "roc_auc": roc_auc_score(yTest, yProbs)
    })
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
with mlflow.start_run(run_name="rf_baseline"):
    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        random_state=42,
        class_weight="balanced"
    )

    pipeline = Pipeline(steps = [
        ("preprocessing", preprocessingNonLinear),
        ("model", model)
    ])

    pipeline.fit(xTrain, yTrain)

    yPredictions = pipeline.predict(xTest)
    yProbs = pipeline.predict_proba(xTest)[:, 1]

    mlflow.log_params(model.get_params())
    
    mlflow.log_metrics({
        "accuracy": accuracy_score(yTest, yPredictions),
        "precision": precision_score(yTest, yPredictions),
        "recall": recall_score(yTest, yPredictions), 
        "f1": f1_score(yTest, yPredictions),
        "roc_auc": roc_auc_score(yTest, yProbs)
    })
```
